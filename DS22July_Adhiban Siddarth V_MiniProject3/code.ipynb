The data set is same as the previous. The task is to apply advanced techniques like Random Forest, AdaBoost and XGBoost on the dataset and compare the performances with the basic methods.

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Dry_Bean.csv')
df
```

```python
df.info()
```

```python
df.Class.unique()
```

## Converting `Class` to Numeric

```python
df.Class = df.Class.map({'SEKER':0, 'BARBUNYA':1, 'BOMBAY':2, 'CALI':3, 'HOROZ':4, 'SIRA':5, 'DERMASON':6})
```

```python
df.head()
```

## Spliting dataset into training and testing datasets

```python
x = df.drop('Class', axis=1).values
y = df['Class'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Creating Machine Learning Models

```python
Rf = make_pipeline(StandardScaler(), RandomForestClassifier(n_estimators=500, random_state=0))
AdaB = make_pipeline(StandardScaler(), AdaBoostClassifier(n_estimators=500, random_state=0))
XgB = make_pipeline(StandardScaler(), XGBClassifier(n_estimators=500, random_state=0))
```

## Training the models

```python
Rf.fit(x_train, y_train)
AdaB.fit(x_train, y_train)
XgB.fit(x_train, y_train)
```

## Predicting the y value for x_test

```python
Rf_predict = Rf.predict(x_test)
AdaB_predict = AdaB.predict(x_test)
XgB_predict = XgB.predict(x_test)
```

## Comparing the best Model

```python
temp = []
for name, predction in zip(['RandomForestClassifier', 'AdaBoostClassifier', 'XGBClassifier'], [Rf_predict, AdaB_predict, XgB_predict]):
    temp.append([name, accuracy_score(y_test, predction)])
result = pd.DataFrame(temp, columns=['Model', 'Accuracy']).sort_values('Accuracy', ascending=False).reset_index(drop=True)
result
```

## **Conclusion:** XGBClassifier gives more Accuracy.
