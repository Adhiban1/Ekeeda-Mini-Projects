```python
import pandas as pd
```

## importing dataset

```python
df = pd.read_csv('Sales_Transactions_Dataset_Weekly.csv')
df
```

```python
df.info()
```

- `float64(52), int64(54), object(1)` only one columns is not in numeric, so drop this column

```python
df1 = df.drop('Product_Code', axis=1)
```

```python
df1
```

## import KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df1)
labels = kmeans.predict(df1)
centroids = kmeans.cluster_centers_
```

```python
labels
```

```python
centroids
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
df2 = pca.fit_transform(df1)
```

```python
df2
```

```python
kmeans1 = KMeans(n_clusters=3)
kmeans1.fit(df2)
labels1 = kmeans1.predict(df2)
centroids1 = kmeans1.cluster_centers_
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x=df2[:, 0], y=df2[:, 1], hue=labels1);
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
model = make_pipeline(StandardScaler(), 
                      PCA(n_components=2), 
                      KMeans(n_clusters=3))

model.fit(df1)
labels2 = model.predict(df1)
centroids2 = model[2].cluster_centers_
```

```python
df3 = model['pca'].transform(df1)
```

```python
sns.scatterplot(x=df2[:, 0], y=df2[:, 1], hue=labels2);
```

```python
sns.scatterplot(x=df3[:, 0], y=df3[:, 1], hue=labels2);
```
